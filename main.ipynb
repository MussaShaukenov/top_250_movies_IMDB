```python
import pandas as pd

df = pd.read_csv('IMDB Top 250 Movies.csv', usecols=['rank', 'name', 'year', 'genre', 'rating'])
df
```

```python
less_2000 = df.loc[df['year'] < 2000]
less_2000
```

```python
most_popular_genre_before_2000 = less_2000['genre'].mode()
most_popular_genre_before_2000
```

```python
after_2000 = df.loc[df['year'] >= 2000]
after_2000
```

```python
most_popular_genre_after_2000 = after_2000['genre'].mode()
print(most_popular_genre_after_2000)
```

```python
number_of_movies_before_2000_among_top_250 = less_2000.shape[0]
number_of_movies_before_2000_among_top_250 / 250 * 100
```

```python
number_of_movies_after_2000_among_top_250 = after_2000.shape[0]
number_of_movies_after_2000_among_top_250 / 250 * 100
```

```python
import matplotlib.pyplot as plt

x = df['year']
y = df['rating']

plt.scatter(x, y)
```

```python
sizes = [less_2000.shape[0], after_2000.shape[0]]
labels = ['Movies before 2000 (61.6 %)', 'Movies after 2000 (38.4 %)']
colors = ['#30053A', '#E6681B']

plt.title('Comparison of movies filmed before and after 2000s')
patches, texts = plt.pie(sizes, labels=labels, colors=colors, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
```
